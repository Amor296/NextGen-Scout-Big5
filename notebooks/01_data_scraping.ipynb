```python
import pandas as pd

# المسار حسب الهيكل بتاعك
file_path = '../data/raw_big5_stats.csv'

try:
    # 1. محاولة القراءة مع تجاهل الأخطاء تماماً
    # جربنا نغير المحرك لـ 'python' عشان يلم الـ 1017 لاعب بتوعك صح
    df = pd.read_csv(file_path, on_bad_lines='warn', engine='python', sep=None)
    
    # 2. لو لقيت الداتا لسه "None"، ده معناه إن الملف محتاج تنظيف يدوي سريع
    # هنشيل الصفوف الفاضية (NaN) اللي ظهرت عندك في النتيجة
    df = df.dropna(how='all').reset_index(drop=True)
    
    print(f"✅ المرة دي بجد.. معانا {len(df)} سطر بيانات.")
    print("إليك أول 5 لعيبة بعد التنظيف:")
    print(df.iloc[:5, :5]) # بنعرض أول 5 صفوف وأول 5 أعمدة

except Exception as e:
    print(f"❌ لسه فيه شوشرة في الإشارة: {e}")
```
