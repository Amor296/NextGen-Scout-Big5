## Phase 1: Data Extraction & Initial Formatting
Coach's Note: In this step, we are identifying our key players from the raw data and making sure everyone is in their right position.

```python
import pandas as pd

# Define the file path within the project structure
file_path = '../data/raw_big5_state.csv'

try:
    # 1. Read the raw file without predefined headers to inspect structure
    # Using 'latin1' to handle special characters found in European player names
    df_raw = pd.read_csv(file_path, header=None, encoding='latin1', on_bad_lines='skip')
    
    # 2. Clean the pitch: Remove columns that are entirely empty
    df_raw = df_raw.dropna(how='all', axis=1) 
    
    # 3. Tactical Positioning: Map specific columns based on FBRef standard layout
    df = pd.DataFrame()
    df['Player'] = df_raw.iloc[:, 1] # Usually the 2nd column contains Player Names
    df['Squad']  = df_raw.iloc[:, 4] # Usually the 5th column contains Team Names
    df['Age']    = df_raw.iloc[:, 6] # Usually the 7th column contains Age (Format: YY-DDD)
    df['Gls']    = df_raw.iloc[:, 12]# Usually the 13th column contains Goals scored
    
    # 4. Filter Non-Players: Ensure the 'Player' column contains actual text (names)
    # We remove rows that are just numbers or contain header text like 'Player' or 'nan'
    df['Player'] = df['Player'].astype(str)
    df = df[df['Player'].str.contains('[a-zA-Z]', na=False)] 
    df = df[~df['Player'].str.contains('Player|None|nan', na=False)]
    
    # 5. Success Message: Report the number of valid players discovered
    print(f"✅ Path Corrected! We have {len(df)} real players in the squad.")
    print(df.head(10))
    
    # 6. Save the Scouting Report: Export the clean data for the next phase
    df.to_csv('../data/final_cleaned_stats.csv', index=False)

except Exception as e:
    print(f"❌ Tactical Error during execution: {e}")
```
